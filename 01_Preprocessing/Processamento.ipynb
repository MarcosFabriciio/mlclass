<h1> Teste Mann-Whitney para gravidez </h1>

```python
lPreg0 = [];
lPreg1 = [];
n = 0;
for val in df.iloc[:,8]:
    if val == 0:
        lPreg0.append(df.iloc[n, 0]);
    else:
        lPreg1.append(df.iloc[n, 0]);
    n = n+1;
print(lPreg0);
print(lPreg1);
```

```python
print(len(lPreg0));
print(len(lPreg1));
print(len(lPreg0) + len(lPreg1));
```

```python
stats.ttest_ind(lPreg0, lPreg1)
```

```python
print(sum(lPreg0)/len(lPreg0));
```

```python
print(sum(lPreg1)/len(lPreg1))
```

```python
stats.mannwhitneyu(lPreg0, lPreg1, False, 'two-sided')
```

<h1> Teste T de Glicose </h1>

```python
lGluc0 = [];
lGluc1 = [];
n = 0;
for val in df.iloc[:,8]:
    if val == 0 and not np.isnan(df.iloc[n, 1]):
        lGluc0.append(df.iloc[n, 1]);
    elif val == 1 and not np.isnan(df.iloc[n, 1]):
        lGluc1.append(df.iloc[n, 1]);
    n = n+1;
print(lGluc0);
print(lGluc1);
```

```python
print(len(lGluc0));
print(len(lGluc1));
print(len(lGluc0) + len(lGluc1));
```

```python
print(sum(lGluc0)/len(lGluc0));
```

```python
print(sum(lGluc1)/len(lGluc1));
```

```python
print(stats.ttest_ind(lGluc0, lGluc1));
```

<h1> Teste T para Pressão Arterial </h1>

```python
lPress0 = [];
lPress1 = [];
n = 0;
for val in df.iloc[:,8]:
    if val == 0 and not np.isnan(df.iloc[n, 2]):
        lPress0.append(df.iloc[n, 2]);
    elif val == 1 and not np.isnan(df.iloc[n, 2]):
        lPress1.append(df.iloc[n, 2]);
    n = n+1;
print(lPress0);
print(lPress1);
```

```python
print(len(lPress0));
print(len(lPress1));
print(len(lPress0) + len(lPress1));
```

```python
print(sum(lPress0)/len(lPress0));
```

```python
print(sum(lPress1)/len(lPress1));
```

```python
print(stats.ttest_ind(lPress0, lPress1));
```
