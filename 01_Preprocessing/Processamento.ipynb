<h1> Teste Mann-Whitney para gravidez </h1>

```python
lPreg0 = [];
lPreg1 = [];
n = 0;
for val in df.iloc[:,8]:
    if val == 0:
        lPreg0.append(df.iloc[n, 0]);
    else:
        lPreg1.append(df.iloc[n, 0]);
    n = n+1;
print(lPreg0);
print(lPreg1);
```

```python
print(len(lPreg0));
print(len(lPreg1));
print(len(lPreg0) + len(lPreg1));
```

```python
stats.ttest_ind(lPreg0, lPreg1)
```

```python
print(sum(lPreg0)/len(lPreg0));
```

```python
print(sum(lPreg1)/len(lPreg1))
```

```python
stats.mannwhitneyu(lPreg0, lPreg1, False, 'two-sided')
```

```python
n = 0;
for val in df['Glucose']:
    if np.isnan(val):
        dfLimpo.at[n, 'Glucose'] = 113;
    n = n+1;
print(dfLimpo)
```

```python
n = 0;
s = 0;
for val in dfLimpo['BMI']:
    if not np.isnan(val):
        n = n+1;
        s = s+val;
mediaBMI = round(s/n, 1);
print(mediaBMI);
```

```python
n = 0;
s = 0;
for val in dfLimpo['BloodPressure']:
    if not np.isnan(val):
        n = n+1;
        s = s+val;
mediaBP = round((s/n));
print(mediaBP);
```

```python
dfLimpo.index = range(len(dfLimpo))
```

```python
n = 0;
m = 0;
for val in dfLimpo['BMI']:
    if np.isnan(val):
        dfLimpo.at[n, 'BMI'] = mediaBMI;
    n = n+1;
for val in dfLimpo['BloodPressure']:
    if np.isnan(val):
        dfLimpo.at[m, 'BloodPressure'] = mediaBP;
    m = m+1;    

print(dfLimpo)
```
